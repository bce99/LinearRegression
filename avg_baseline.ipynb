```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error, root_mean_squared_error
```

```python
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
X = df.drop('Outcome', axis=1)
y = df[['Outcome']]
feats = len(X.columns)
```

```python
# Check for label proportions

print("Percentage of labels with no diabetes:", round(len(df[df['Outcome'] == 0]) / len(df) * 100, 2), "%")
print("Percentage of labels with diabetes:", round(len(df[df['Outcome'] == 1]) / len(df) * 100, 2), "%")
```

```python
y_hat = np.zeros(len(df))
```

```python
mse  = mean_squared_error(y, y_hat)
rmse = root_mean_squared_error(y, y_hat)  # RMSE
mae  = mean_absolute_error(y, y_hat)
r2   = r2_score(y, y_hat)
adj_r2 = 1-(1-r2_score(y, y_hat)) * (len(X)-1) / (len(X) - feats - 1)
mape = mean_absolute_percentage_error(y, y_hat) * 100

print("MSE :", mse)
print("RMSE:", rmse)
print("MAE :", mae)
print("R²  :", r2)
print("Adj R²:", adj_r2)
print("MAPE:", mape, "%")
```
