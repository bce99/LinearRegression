```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error, root_mean_squared_error
```

```python
df = pd.read_csv('diabetes.csv')
df.head()
```

Brute force solve using LR

```python
y = np.array(df['Outcome'])

X = df.drop('Outcome', axis=1).to_numpy()
feats = df.drop('Outcome', axis=1).columns.to_list()

# Add a column of 1 for intercept
X_int = np.column_stack([np.ones(len(X)), X])
```

Recall that:
$$
\hat{\beta} = \left( X^\mathsf{T} X \right)^{-1} X^\mathsf{T} y
$$


```python
def solve_normal(X, y):
    XtX = X.T @ X
    Xty = X.T @ y
    beta = np.linalg.solve(XtX, Xty)

    return beta, XtX, Xty
```

```python
beta, XtX, Xty = solve_normal(X_int, y)
```

```python
y_hat = X_int @ beta

y_hat = np.where(np.abs(y_hat-0) < np.abs(y_hat-1), 0, 1)

mse  = mean_squared_error(y, y_hat)
rmse = root_mean_squared_error(y, y_hat)  # RMSE
mae  = mean_absolute_error(y, y_hat)
r2   = r2_score(y, y_hat)
adj_r2 = 1-(1-r2_score(y, y_hat)) * (len(X_int)-1) / (len(X_int) - len(feats) - 1)
mape = mean_absolute_percentage_error(y, y_hat) * 100

print("MSE :", mse)
print("RMSE:", rmse)
print("MAE :", mae)
print("R²  :", r2)
print("Adj R²:", adj_r2)
print("MAPE:", mape, "%")
```
